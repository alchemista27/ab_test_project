```python
# Import the libary
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as GridSpec
import datetime
import dateutil


# Load the data
data = pd.read_csv('Testdata.csv')
data.head()
```

```python
# Checking the duplicated value
data['id_user'].duplicated().sum()

data = data.drop('id_user', axis=1)
```

```python
# filling up the missing value
data['paid_plan_signup_date'] = data['paid_plan_signup_date'].fillna(0)
data.head()


# Creating the signup columns
signup = []
for x in data['paid_plan_signup_date']:
    if x != 0:
        signup.append(1)
        
    else:
        signup.append(0)
        
data['signup_or_not'] = signup

# Change the treatment columns to category
data[['treatment', 'platform', 'time_zone']] = data[['treatment', 'platform', 'time_zone']].astype('category')
data.info()
```

```python
sns.barplot(data = data, x = 'treatment', y = 'signup_or_not', errorbar=('ci', False))
plt.xlabel('Group of Experiment')
plt.ylabel('Total Signup for Paid Plan')
plt.title('Comparisson of total signup between two groups experiment')
plt.savefig('Group vs Signup.png')
plt.show()

# Slighly there is an indication that the campaign impacted that more customer sign up for the paid plan
```

```python
sns.barplot(data = data, x = 'platform', y = 'signup_or_not', errorbar=('ci', False))
plt.title('Comparison of total signup between platform used')
plt.xlabel('Type of Platform')
plt.ylabel('Total Signup for Paid Plan')
plt.savefig('Platform vs Signup.png')
plt.show()
```

```python
sns.barplot(data = data, x = 'time_zone', y = 'signup_or_not', errorbar=('ci', False))
plt.title('Comparison of total signup between time zone of the user')
plt.xlabel('User Time Zone')
plt.ylabel('Total Signup for Paid Plan')
plt.savefig('Time Zone vs Signup.png')
plt.show()
```

```python
# Spliting the data to control data and treatment data
control = data[data['treatment']==0]
treatment = data[data['treatment']==1]

# Normality check for both control group
from scipy.stats import normaltest

alpha = 0.05
stats, p_value = normaltest(control['signup_or_not'])
print(p_value)
if p_value < alpha:
    print('The data of the control group is not Normally distributed')
else:
    print('The data of the control group is Normally distributed')
    
    
# Normality check for treatment group
stats, p_value1 = normaltest(treatment['signup_or_not'])
print(p_value1)
if p_value1 < alpha:
    print('The data of the treatment group is not Normally distributed')
else:
    print('The data of the treatment group is Normally distributed')

# The data is not normally distributed, we will use non parametric testing for ab test
from scipy.stats import mannwhitneyu

res, p_value3 = mannwhitneyu(treatment['signup_or_not'], control['signup_or_not'], alternative='greater')

print(p_value3)
if p_value3 < alpha:
    print('The Control Group is significantly less than the treatment group')
else:
    print('The Treatment Group is significantly greater than Control Group')
```

```python
# Selecting data of customers who signup of paid plan
data_signup = data[data['paid_plan_signup_date']!=0]

# converting the value in experiment_date and paid_plan_signup_date into datetime
cols = ['experiment_date', 'paid_plan_signup_date']
for col in cols:
    data_signup[col] = pd.to_datetime(data_signup[col])
    
data_signup.head()
```

```python
# calculating timedelta
data_signup['timedelta'] = data_signup['paid_plan_signup_date'] - data_signup['experiment_date']
data_signup

# Convert the timedelta into float64
data_signup['timedelta'] = data_signup['timedelta'].astype('int64')
data_signup.head()

# Selecting the data for normality test
control = data_signup[data_signup['treatment']==0]
treatment = data_signup[data_signup['treatment']==1]

# Checking the normality
stats, p_value = normaltest(control['timedelta'])
print(p_value)
if p_value < alpha:
    print('The data of the control group is not Normally distributed')
else:
    print('The data of the control group is Normally distributed')
    
stats, p_value = normaltest(treatment['timedelta'])
print(p_value)
if p_value < alpha:
    print('The data of the control group is not Normally distributed')
else:
    print('The data of the control group is Normally distributed')
    
    
stats, p_value = mannwhitneyu(control['timedelta'], treatment['timedelta'], alternative='greater')
print(p_value)
if p_value < alpha:
    print('The timedelta of control group is less than treatment group')
else:
    print('The timedelta of control group is greater than treatment group')
    
# Visualize the timedelta and customer who cancel
sns.barplot(data=data_signup, x='treatment', y='timedelta', errorbar=('ci', False))
plt.xlabel('Group of Experiments')
plt.ylabel('Total time spend before purchasing')
plt.title('Comparison of total spend vs group of experiment')
plt.savefig('Group vs Total time.png')
plt.show()
```

```python
sns.barplot(data=data_signup, x='treatment', y='cancelled_paid_plan_or_not', errorbar=('ci', False))
plt.xlabel('Group of experiment')
plt.ylabel('Total user cancelled the paid plan')
plt.title('Comparison between group and user who cancelled')
plt.savefig('Group vs cancelled.png')
plt.show()
```

```python
# Checking the difference of treatment and control cancel rate
# Checking the normality
stats, p_value = normaltest(control['cancelled_paid_plan_or_not'])
print(p_value)
if p_value < alpha:
    print('The data of the control group is not Normally distributed')
else:
    print('The data of the control group is Normally distributed')
    
stats, p_value = normaltest(treatment['cancelled_paid_plan_or_not'])
print(p_value)
if p_value < alpha:
    print('The data of the control group is not Normally distributed')
else:
    print('The data of the control group is Normally distributed')
    
stats, p_value = mannwhitneyu(control['cancelled_paid_plan_or_not'], treatment['cancelled_paid_plan_or_not'], alternative='greater')
print(p_value)
if p_value < alpha:
    print('The cancel rate of control group is less than treatment group')
else:
    print('The cancel rate of control group is greater than treatment group')
```
